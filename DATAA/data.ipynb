```python
import pandas as pd
import matplotlib.pyplot as plt
```

1. Data loading


```python
data = pd.read_csv('hardwareStore.csv')
data.head()
```

2. Initial Data Analysis

```python
data = pd.read_csv('hardwareStore.csv', encoding = "latin-1")
data.head()
```

```python
Devices = data.groupby('CATEGORY_NAME')['QUANTITY'].sum()


std_cost_mean = data['STANDARD_COST'].mean()
std_cost_median = data['STANDARD_COST'].median()
std_cost_std = data['STANDARD_COST'].std()
print("STANDARD COST")
print("mean = " , standard_cost_mean)
print("median = " , standard_cost_median)
print("STANDARD DEVIATION = " , standard_cost_std)
print(" ")
std_cost_mean1 = data['LIST_PRICE'].mean()
std_cost_median1 = data['LIST_PRICE'].median()
std_cost_std1 = data['LIST_PRICE'].std()
print("LIST PRICE")
print("mean = " , standard_cost_mean1)
print("median = " , standard_cost_median1)
print("STANDARD DEVIATION = " , standard_cost_std1)
print(" ")
std_cost_mean12 = data['QUANTITY'].mean()
std_cost_median12 = data['QUANTITY'].median()
std_cost_std12 = data['QUANTITY'].std()
print("QUANTITY")
print("mean = " , standard_cost_mean12)
print("median = " , standard_cost_median12)
print("STANDARD DEVIATION = " , standard_cost_std12)
```

3. Category Analysis

```python
Devices.plot(kind = 'bar')
plt.xlabel('')
plt.ylabel('the number of products')
plt.title('products')
plt.xticks(rotation = 40)
plt.show
```

4. Cost and Price Analysis:

```python
ave_standard_cost = data['STANDARD_COST'].mean()
ave_list_price = data['LIST_PRICE'].mean()
highest_standard_cost_product = data.loc[data['STANDARD_COST'].idxmax(), 'PRODUCT_NAME']
highest_list_price_product = data.loc[data['LIST_PRICE'].idxmax(), 'PRODUCT_NAME']
print("The Average Standard Cost:", ave_standard_cost)
print("The Average List Price:", ave_list_price)
print("The Product with Highest Standard Cost:", highest_standard_cost_product)
print("The Product with Highest List Price:", highest_list_price_product)
```

5. Location Analysis:

```python
loc_quantity = data.groupby('STATE')['QUANTITY'].sum()
highest_quantity_location = loc_quantity.idxmax()
print("The Location with the Highest Quantity of Products in Stock is:", highest_quantity_location)
```
