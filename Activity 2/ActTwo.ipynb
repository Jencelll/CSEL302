```python
import pandas as pd
import matplotlib.pyplot as plt
```

#1: Data Loading


```python
hardware_store_df = pd.read_csv("hardwareStore.csv")
```

#2: Initial Data Analysis

```python
print("Basic Statistics of Numeric Columns:")
print(hardware_store_df.describe())
print("\nUnique Categories and Their Counts:")
print(hardware_store_df['CATEGORY_NAME'].value_counts())
```

#3: Category Analysis

```python
category_counts = hardware_store_df['CATEGORY_NAME'].value_counts()
category_counts.plot(kind='bar', xlabel='Category', ylabel='Number of Products', title='Distribution of Products Across Categories')
plt.xticks(rotation=45)
plt.show()

highest_category = category_counts.idxmax()
print("\nCategory with the Highest Number of Products:", highest_category)
```

#4: Cost and Price Analysis

```python
average_standard_cost = hardware_store_df['STANDARD_COST'].mean()
average_list_price = hardware_store_df['LIST_PRICE'].mean()
print("\nAverage Standard Cost:", average_standard_cost)
print("Average List Price:", average_list_price)

highest_standard_cost_product = hardware_store_df.loc[hardware_store_df['STANDARD_COST'].idxmax(), 'PRODUCT_NAME']
highest_list_price_product = hardware_store_df.loc[hardware_store_df['LIST_PRICE'].idxmax(), 'PRODUCT_NAME']
print("Product with the Highest Standard Cost:", highest_standard_cost_product)
print("Product with the Highest List Price:", highest_list_price_product)
```

#5: Location Analysis

```python
highest_quantity_location = hardware_store_df.loc[hardware_store_df['QUANTITY'].idxmax(), 'CITY']
print("\nLocation with the Highest Quantity:", highest_quantity_location)
```
