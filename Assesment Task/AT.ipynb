```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('Lesson_4_student_grades_dataset 2.csv')
data.head()
```

```python
Hours_of_Study = data.drop('Grades', axis=1)
Grades = data[['Grades']]
```

```python
Hours_of_Study_train, Hours_of_Study_test, Grades_train, Grades_test = train_test_split(Hours_of_Study, Grades, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(Hours_of_Study_train, Grades_train)
```

```python
plt.scatter(data['Hours_of_Study'],data['Grades'])
plt.xlabel('Hours of study (X)')
plt.ylabel('Grades (Y)')
plt.title('')
plt.show()
```

```python
plt.scatter(data['Hours_of_Study'],data['Grades'])
plt.plot(Hours_of_Study,model.predict(Hours_of_Study), color='red')
plt.xlabel('Hours of study')
plt.ylabel('Grades')
plt.title('Linear Regression')
plt.xticks(rotation=5)
plt.show()
```

```python
new_data = np.array([[25,65,76,2,47]])

prediction = model.predict(new_data)

print(f"Predicted Grade: {prediction[0]}")
```
